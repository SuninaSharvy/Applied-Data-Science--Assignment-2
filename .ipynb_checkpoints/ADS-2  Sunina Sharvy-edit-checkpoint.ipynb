##### Defining Functions

```python
#Funtions used for this projects mentioned below

def World_Bank_Reader (wb_file_name):
    """Doc String """
    
    #Reading data from the world bank csv file downloaded from the world bank website
    cchange_df = pd.read_csv(wb_file_name, skiprows = 4) 
    cchange_df.drop(cchange_df.columns[[-1]], axis=1, inplace=True)
    year_df = cchange_df[:]  #Deriving data with Years on the columns
    
    # Deriving dataframe with country name as columns
    country_df = cchange_df.groupby(['Country Name']).aggregate('first').transpose()
    
    return country_df,year_df


def heat_map (data):
    
    """Doc String """
    
    plt.figure(figsize=(14,8)) # Adjusting the figure size
    sns.heatmap(data,annot = True, cmap="YlGnBu")  # Using .heatmap funtion from the SNS library for ploting heat map
    plt.show()

    
def WB_Time_Series_Plt(wb_data, Y_Indicator,xlbl,ylbl,plt_title,colour):
    
    """Doc String """
    
    plt.rcParams['figure.figsize']= (6,6)  #Setting the figure size
    time_data = wb_data[:]                 #Assigning data attributes to the local variable
    
    time_data = time_data[time_data['Country Code'] != 'World'] # Removing the World information data to process the data easily
    time_data = time_data.groupby(['Indicator Code']).aggregate('mean').transpose().fillna(0) #Grouping data using Indicator code
    time_data = time_data[[Y_Indicator]]   # Saving Timing data derived usign Indicator as the index
    time_data = time_data.reset_index()    #Resetting the Index to normal
    time_data = time_data.rename(columns = {'index':'Year'}) #Renaming the Data frame Index as an year
    time_data.Year = time_data.Year.values.astype(int)
    plt.plot(time_data['Year'], time_data[Y_Indicator],color=colour) #ploting the Time series
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    plt.title(plt_title)
    plt.show()
    
    
def World_Bank_Single_Reader (Indicator,country_codes,start_year,end_year):
    
    """Doc String """
    #It is used for accessing single indicator data from World Bank website (another method)
    cchange_df = wb.data.DataFrame(Indicator,country_codes, time=range(start_year, end_year))
    country_df= cchange_df.transpose()  #Transpose To get the Year Values as a main column
    country_df.index.names = ["Years"]

    return country_df


def Bar_Plt_Multi (data):
    
    """Doc String """
    
    #Setting up the axis information to fit on the Bar plot
    ax = data.unstack(level=0).plot(kind='bar', subplots=True, rot=90, figsize=(18, 14), layout=(4, 2))
    plt.tight_layout()

    
def histogram_plt ( data,country_code,indicator,xlbl,plt_title,colour):
    
    """Doc String """
    #ploting the Histogram using the data frame & corresponding indicator
    plt.rcParams['figure.figsize']= (5,5)
    data_1=data[country_code]
    data_2 = data_1[indicator]
    plt.hist(data_2, bins=15,color = colour)
    
    plt.xlabel(xlbl)
    plt.ylabel("Count")
    plt.title(plt_title)
    plt.show()
    

def sort(sort_data):
    """Doc String """
    #Used for Sorting the data frame with Median.
    return sort_data.sort_values(ascending=False, by='median')
```
